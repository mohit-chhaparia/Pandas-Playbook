```python
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame(
    {
        "order_id": ["A101" , "A102" , "A103" , "A104" , "A105"] ,
        "region": ["East" , "West" , "East" , "West" , "East"] ,
        "amount": [250 , 900 , 120 , 650 , 480] ,
        "status": ["Delivered" , "Delivered" , "Pending" , "Delivered" , "Pending"] ,
    } ,
    index = pd.to_datetime([
        "2025-12-01" , "2025-12-02" , "2025-12-03" , "2025-12-04" , "2025-12-05"
    ]) , )
df
```

```python
#===============================
# Case 1) .loc[] single label
#===============================
case1 = df.loc["2025-12-02" , ["order_id" , "amount"]]
case1
```

```python
#===============================
# Case 2) .loc[] list of labels
#===============================
case2 = df.loc[["2025-12-01" , "2025-12-04"] , ["order_id" , "region" , "amount"]]
case2
```

```python
#============================================
# Case 3) .loc[] label slice (inclusive end)
#============================================
case3 = df.loc["2025-12-02":"2025-12-04" , ["order_id" , "amount"]]
case3
```

```python
#=================================
# Case 4) .loc[] boolean filter
#=================================
case4 = df.loc[df["amount"] >= 500 , ["order_id" , "amount" , "status"]]
case4
```

```python
#====================================
# Case 5) .loc[] callable (lambda)
#====================================
case5 = df.loc[lambda x: x["status"].eq("Delivered") , ["order_id" , "region" , "amount"]]
case5
```

```python
#======================================
# Case 6) .iloc[] positions (row/col)
#======================================
case6 = df.iloc[0 , :]
case6

case6b = df.iloc[0 , 2]
case6b
```

```python
#========================================
# Case 7) .iloc[] slice (stop excluded)
#========================================
case7 = df.iloc[0:3 , 0:3]
case7
```

```python
#======================================
# Case 8) .iloc[] lists of positions
#======================================
case8 = df.iloc[[3 , 1] , [0 , 2 , 3]]
case8
```
