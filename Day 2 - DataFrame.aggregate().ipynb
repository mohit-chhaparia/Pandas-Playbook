```python
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({
    "region":   ["East","East","West","West"],
    "sales":    [120, 80, 200, 60],
    "returns":  [5, 0, 8, 2],
    "discount": [0.10, 0.05, 0.00, 0.15]
})
df
```

```python
# Case 1
num = df[["sales","returns","discount"]]
num.agg(lambda s: s.max() - s.min())   # range per column
```

```python
# Case 2
df[["sales","returns"]].agg("sum")
```

```python
# Case 3
df[["sales","returns"]].agg(["min", "max", "mean"])
```

```python
# Case 4
df.agg({
    "sales": ["sum", "mean"],
    "returns": "max",
    "discount": "mean"
})
```

```python
# Case 5
df[["sales"]].agg(
    total_sales=("sales","sum"),
    avg_sales=("sales","mean"),
    best_sale=("sales","max")
)
```

```python
# Case 6
scores = pd.DataFrame({"math":[80,90,60], "eng":[70,85,75], "sci":[75,95,65]})
scores
scores.agg("mean", axis="columns")   # row-wise mean
```
